Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.

```python
import seaborn as sns
import plotly.express as px

# Load the dataset
titanic = sns.load_dataset("titanic")

# Plot scatter plot using Plotly Express
fig = px.scatter(titanic, x='age', y='fare', title='Scatter Plot of Age vs Fare in Titanic Dataset')
fig.show()
```

Explanation:

sns.load_dataset("titanic"): Loads the "titanic" dataset from Seaborn. px.scatter(): Creates a scatter plot using Plotly Express with age on the x-axis and fare on the y-axis.

Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.

```python
import plotly.express as px
df = px.data.tips()

# Plot histogram using Plotly Express
fig = px.histogram(df, x='sex', y='total_bill', color='day', barmode='group',
                   pattern_shape='smoker', title='Histogram of Total Bill by Sex with Smoker Pattern')
fig.show()
```

Explanation:

px.histogram(): Creates a histogram with sex on the x-axis and total_bill on the y-axis, grouped by day. pattern_shape='smoker': Uses the "smoker" column to differentiate groups with different pattern shapes. color='day': Colors bars based on the "day" column.

Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for the color parameter. Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the dimensions parameter.

```python
import plotly.express as px
from sklearn.datasets import load_iris
import pandas as pd

# Load the iris dataset
iris_sklearn = load_iris()
iris_df = pd.DataFrame(data=iris_sklearn.data, columns=iris_sklearn.feature_names)
iris_df['species'] = iris_sklearn.target_names[iris_sklearn.target]

# Plot scatter matrix plot using Plotly Express
fig = px.scatter_matrix(iris_df, dimensions=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
                        color='species', title='Scatter Matrix Plot of Iris Dataset')
fig.show()
```

Q5. What is Distplot? Using Plotly express, plot a distplot. A distplot (distribution plot) in Plotly Express combines a histogram with a kernel density estimate (KDE) plot. It displays the distribution of a continuous variable.

```python
import plotly.express as px
import numpy as np

# Generate sample data
np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=1000)

# Plot distplot using Plotly Express
fig = px.histogram(data, x=data, nbins=30, marginal='rug', title='Distplot of Random Data')
fig.show()
```

Explanation:

px.histogram(): Creates a histogram (distplot) of data, displaying the distribution with a KDE. marginal='rug': Adds marginal rug plots to show individual data points.
