Q2: What is a scatter plot? Use the given data to plot a scatter plot.

Scatter Plot:

 - A scatter plot is a type of plot that displays values for two variables as points on a Cartesian plane, where each       point represents an observation in the dataset.

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

Explanation:

 - plt.scatter(x, y): Plots a scatter plot using x as the x-axis data and y as the y-axis data.
- plt.title(), plt.xlabel(), plt.ylabel(): Add title, x-axis label, and y-axis label respectively to the plot.

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

subplot() Function:

- subplot() is used to create multiple plots (subplots) within the same figure.
- It allows you to arrange plots in a grid format specified by rows and columns.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for four lines
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Creating subplots
plt.figure(figsize=(12, 8))

# Subplot 1
plt.subplot(221)  # 2 rows, 2 columns, plot 1
plt.plot(x, y1)
plt.title('Line Plot 1')

# Subplot 2
plt.subplot(222)  # 2 rows, 2 columns, plot 2
plt.plot(x, y2)
plt.title('Line Plot 2')

# Subplot 3
plt.subplot(223)  # 2 rows, 2 columns, plot 3
plt.plot(x, y3)
plt.title('Line Plot 3')

# Subplot 4
plt.subplot(224)  # 2 rows, 2 columns, plot 4
plt.plot(x, y4)
plt.title('Line Plot 4')

plt.tight_layout()
plt.show()
```

Explanation:

- plt.subplot(221), plt.subplot(222), plt.subplot(223), plt.subplot(224): Defines the position of each subplot within a   2x2 grid.
- plt.plot(x, y): Plots the line for each subplot with data x and y.
 - plt.tight_layout(): Adjusts subplot parameters to give specified padding.

Q4: What is a bar plot? Why is it used? Plot a bar plot and a horizontal bar plot using the given data.

Bar Plot:

- A bar plot is a type of plot that presents categorical data with rectangular bars with lengths proportional to the values they represent.
 - It is used to compare values of different categories or track changes over time.

```python
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Plotting a bar plot
plt.figure(figsize=(8, 5))
plt.bar(company, profit, color='skyblue')
plt.title('Bar Plot: Company vs Profit')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.show()

# Plotting a horizontal bar plot
plt.figure(figsize=(8, 5))
plt.barh(company, profit, color='lightgreen')
plt.title('Horizontal Bar Plot: Company vs Profit')
plt.xlabel('Profit')
plt.ylabel('Company')
plt.show()
```

Explanation:

- plt.bar(company, profit): Plots a vertical bar plot with company on x-axis and profit on y-axis.
 - plt.barh(company, profit): Plots a horizontal bar plot with company on y-axis and profit on x-axis.
