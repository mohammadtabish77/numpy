Q1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
Plots that can be created using Seaborn:



- Scatter Plot: Shows the relationship between two variables with points that may show a correlation.
  Uses: Useful for exploring correlations and relationships between two continuous variables.

- Line Plot: Displays data points connected by straight line segments, typically used to visualize trends over time.
  Uses: Shows trends and patterns in data, particularly useful for time series data.

- Bar Plot: Represents categorical data with rectangular bars, where the length of each bar is proportional to the values   they represent.
  Uses: Compares quantities across different categories or tracks changes over time.

- Box Plot: Displays the distribution of data based on a five-number summary (minimum, first quartile, median, third quartile, maximum).
  Uses: Identifies outliers and shows the spread and skewness of data distributions.
- Violin Plot: Combines aspects of box plot and kernel density plot, showing the distribution of data across levels of one (or multiple) categorical variables.
  Uses: Provides deeper insights into data distribution compared to box plots, especially useful for comparing distributions across categories.

Q2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
fmri = sns.load_dataset("fmri")

# Plotting line plot for different events and regions
plt.figure(figsize=(10, 6))
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)
plt.title('FMRI Dataset: Signal over Time')
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.show()
```

Explanation:

sns.load_dataset("fmri"): Loads the "fmri" dataset from Seaborn.
sns.lineplot(): Plots a line plot using timepoint on the x-axis and signal on the y-axis, differentiated by event with different line styles for region.

Q3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
titanic = sns.load_dataset("titanic")

# Plotting two box plots
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic)
plt.title('Box Plot of Age by Pclass')

plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic)
plt.title('Box Plot of Fare by Pclass')

plt.tight_layout()
plt.show()
```

Explanation:

sns.load_dataset("titanic"): Loads the "titanic" dataset from Seaborn.
sns.boxplot(): Plots a box plot showing the distribution of age and fare across different passenger classes (pclass).

Q4: Use the "diamonds" dataset from Seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
diamonds = sns.load_dataset("diamonds")

# Plotting histogram with hue
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds, x='price', hue='cut', kde=True)
plt.title('Histogram of Price by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

Explanation:

sns.load_dataset("diamonds"): Loads the "diamonds" dataset from Seaborn.
sns.histplot(): Plots a histogram of price with cut as the hue, showing the distribution of diamond prices across different cuts.
